### Aufgabe 1

Schreiben Sie einen Würfelsimulator. 

- Fragen Sie den Benutzer, wie viele Spieler im Spiel involviert sind.
- Fragen Sie den Benutzer, ob ein Spieler nochmals würfeln darf, wenn eine 6 kommt.

Ablauf:
- Sie zeigen dem Benutzer an, welcher Spieler gerade am Zug ist, 1 ... n
- Sie warten auf eine Eingabe des Benutzers (z.B. Enter), dann wird der Zufallswert zwischen 1 und 6 angezeigt.
- Wenn der Benutzer eine 6 wirft, darf der Benutzer nochmals würfeln, solange, bis keine 6 mehr kommt. Sofern der Benutzer gewählt hat, dass man bei einer 6 nochmals würfeln darf.
- Wenn der Zug beendet ist, zeigen Sie an, dass der nächste Spieler am Zug ist.
- Sie können den Ablauf stoppen, falls statt einer leeren Eingabe z.B. "q" für "quit" eingegeben wird.

```python
#Hier sollte die Lösung stehen

import random
random.seed()

anzahlSpieler = int(input("Bitte geben Sie die Anzahl der Spieler ein: "))
wiederholeWenn6 = input("Soll eine gewürfelte 6 dazu führen, dass weiter gewürfelt werden darf? (y/n)") == "y"

benutzereingabe = ""
while True:
    for spieler in range(anzahlSpieler):
        print("Spieler ", spieler + 1, "ist am Zug! <Enter> um zu würfeln!")
        benutzereingabe = input()
        if benutzereingabe == "q": break #Break inner loop

        while True:
            gewuerfelteZahl = random.randint(1, 6)
            print("Wir würfeln ... ", gewuerfelteZahl)
            if gewuerfelteZahl != 6 or not wiederholeWenn6:
                break
            print("<Enter> um nochmals zu würfeln")
            input() #Waits for user input

    if benutzereingabe == "q": break #Break outer loop
```

```python
#Hier sollte die Lösung stehen
import random
random.seed()

def wuerfle(repeatOn6 : bool) -> bool:
    """Würfelt so lange bis keine 6 mehr kommt (falls repeatOn6 aktiviert). 
       Ansonsten einmal. Liefert zurück, ob der Benutzer abbrechen möchte."""
    while True:
        abbruch = input("Drücken Sie <Enter> um zu würfeln.") == "q"
        x = random.randint(1,6)
        print("Sie würfeln eine", x, "!!!")
        if abbruch or x != 6 or not repeatOn6: return abbruch
        print("Sie dürfen nochmals würfeln!")

def eineRundeWuerfeln(nrSpieler : int, repeatOn6 : bool) -> bool:
    """Würfelt für alle Spieler und liefert zurück, ob der Benutzer 
    das Würfeln abgebrochen hat."""
    for i in range(nrSpieler):
        print("*" * 30)
        print("Hallo Spieler", i + 1)
        abbruch = wuerfle(repeatOn6) or input("!!! Spielerwechsel, bitte bestätigen!") == "q"
        if abbruch: return True
    return False

def start():
    nrSpieler = int(input("Wie viele Spieler sollen spielen? "))
    repeatOn6 = input("Soll bei 6 nochmals gewürfelt werden (y/n)? ") == "y"
    abbruch = False
    while abbruch == False:
        abbruch = eineRundeWuerfeln(nrSpieler, repeatOn6)

start()
```

```python
import random
random.seed()

anzBen = int(input("Bitte Anzahl der Benutzer eingeben: "))
wdbei6 = input("Wollen Sie bei 6 wiederholen (y/n)?") == "y"

def gameLoop():
    while True:
        for pl in range(anzBen):
            aktuelleSpieler = pl + 1
            print("Spieler", aktuelleSpieler, "ist dran.")

            w = 0
            while w == 0 or (w == 6 and wdbei6):
                if input("Bitte Taste drücken (Abbruch? q)") == "q": return
                w = random.randint(1, 6)
                print("Sie würfeln eine", w)

            if input("Wollen Sie aufhören? (y/n)") in ["y", "q"]: return
gameLoop()
```

### Aufgabe 2

Ein Zahlenschloss besteht aus fünf Ringen. Mit je 10 Möglichkeiten. Von 0 bis 9.

Sie haben leider den Code für das Schloss vergessen. Können sich aber an ein paar Einzelheiten erinnern:
- Der Code enthält keine 5.
- Der Code enthält mindestens einmal die Ziffer 3.
- Der Code enthält mindestens einmal die Ziffer 6.
- Der Code startet mit einer geraden Zahl. Aber nicht mit der 0.
- Der Code ist aufsteigend, das heißt, die vorangegangene Zahl ist kleiner oder gleich der nachfolgenden Zahl. 

Schreiben Sie ein Programm, das alle verbleibenden Möglichkeiten ausgibt.

Hinweis: Sie können eine Zahl mit str(zahl) in einen String konvertieren. Mit der Funktion zfill(5) erreichen Sie, dass der String so lange mit führenden "0" aufgefüllt wird, bis dieser die Länge von 5 hat.

    print(str(1).zfill(5)) #führt zur Ausgabe 00001

```python
def isPotentialCodeCandidate(s : str) -> bool:
    if "5" in s: return False             #Code enhält eine 5, dann false
    if "3" not in s: return False         #Code enthält keine 3, dann false
    if "6" not in s: return False         #Code enthält keine 6, dann false
    if s[0] == "0": return False          #Code startet mit 0, dann false
    if int(s[0]) % 2 != 0: return False   #Erste Zahl ist ungerade, dann false
    for i in range(len(s)-1):             #Vergleich der aufsteigenden Ordnung
        if s[i] > s[i+1]: return False    #Falls Ordnung gebrochen, dann false
    return True                           #Alle Prüfungen durch, dann true

    #Alternative Prüfung der Ordnung der einzelnen Zahlen (statt der For-Schleife)
    # if not (s[0] <= s[1] <= s[2] <= s[3] <= s[4]): return False

for comb in range(100000):
    zc = str(comb).zfill(5)
    if isPotentialCodeCandidate(zc): print(zc)
```

```python
#Hier sollte die Lösung stehen

for i in range(100000):
    zahlenschlosszahl = str(i).zfill(5) 

    isCandidate = True

    count3 = 0
    count6 = 0

    #Testen ob 5 in der Zahl vorkommt
    for z in zahlenschlosszahl:
        zahl = int(z)
        if zahl == 5: 
            isCandidate = False
            break
        elif zahl == 3:
            count3 += 1
        elif zahl == 6:
            count6 += 1
    
    if count3 == 0: isCandidate = False
    if count6 == 0: isCandidate = False

    if not isCandidate: continue

    #Startet nicht mit 0 aber mit einer geraden zahl
    if int(zahlenschlosszahl[0]) == 0 or int(zahlenschlosszahl[0]) % 2 != 0: isCandidate = False

    if not isCandidate: continue

    #Für jede Position
    for p in range(len(zahlenschlosszahl) - 1):
        z1 = int(zahlenschlosszahl[p])
        z2 = int(zahlenschlosszahl[p+1])
        if(z1 > z2): #die Vorzahl ist größer als die nachfolgende Zahl
            isCandidate = False
            break

    if not isCandidate: continue

    #An dieser Stelle handelt es sich um einen Kandiat
    print(zahlenschlosszahl)


```
