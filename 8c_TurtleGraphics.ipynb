## Aufgaben Turtle-Graphics

Die *Turtle*-Graphics ist eine schon ältere Möglichkeit, spielend das Programmieren zu lernen. Der Gedanke ist, einem Zeichenstift (dem *Turtle*) anzugeben, wohin er sich bewegen soll. Zusätzlich können Richtung, Farbe, Stiftstärke usw. verändert werden. 

#### Funktionen für Bewegen und Zeichnen
* forward() | fd() : Bewegt den Turtle vorwärts (abhängig davon, wohin er blickt).
* backward() | bk() | back() : Bewegt den Trutle rückwärts.
* right() | rt() : Dreht den Turtle rechts um die angegebenen Grad.
* left() | lt() : Dreht den Turtle links um die angegebenen Grad.
* goto() | setpos() | setposition() : Setzt den Turtle zu der angegebenen Position.
* setx() : Ändert die X-Koordinate des Turtle (Y bleibt unverändert).
* sety() : Ändert die Y-Koordinate des Turtle (X bleibt unverändert).
* setheading() | seth() : Setzt die Orientierung des Turtles in Grad.
* home() : Bewegt den Turtle zurück zu den Anfangskoordinaten (0,0).
* circle(): Zeichnet einen Kreis.
* dot(): Zeichnet einen Punkt.
* undo(): Macht die letzte Änderung rückgängig. Mehrfach ausführbar.
* speed(): Bestimmt die Turtle-Geschwindigkeit (1 langsam, 10 schnell, 0 keine Animation)

#### Funktionen für den Zustand des Turtle
* position() | pos() : Liefert die Position des Turtle.
* towards() : Liefert den Winkel zwischen der Turtle-Position und dem angegebenen Punkt (Vecotr).
* xcor() : Liefert die X-Koordinate des Turtle.
* ycor() : Liefert die Y-Koordinate des Turtle.
* heading(): Liefert den Winkels (im Abhängigkeit zum Koordinaten-System), in dem der Turtle gerade ausgerichtet ist.
* distance() : Liefert die Distanz zwischen dem Turtle und dem angegebenen Punkt.

#### Funktionen um das Zeichnen des Turtles zu kontrollieren
* pendown() | pd() | down() : Der Turtle zeichnet, während er sich bewegt.
* penup() | pu() | up() : Der Turtle zeichnet nicht, während er sich bewegt.
* pensize() | width() : Setzt die Stiftbreite bzw. liefert diese zurück.
* pen() : Eigenschaften des Stifts können gesetzt werden (z.B. shown, pendown, pencolor, fillcolor, pensize, speed, ...).
* isDown() : Liefert zurück, ob der Turtle im Schreibmodus ist.
* color() : Liefert die Linien- und Füllfarbe bzw. lässt diese setzen.
* pencolor() : Setzt die Linien-/Stiftfarbe oder liefert diese zurück.
* fillcolor() : Setzt die Füllfarbe oder liefert diese zurück.
* filling(): LIefert den Füllmodus (True heißt, es wird gefüllt).
* begin_fill() und end_fill() : Aufzurufen, bevor gezeichnet wird. Die gezeichnete Figur wird befüllt, bis end_fill() aufgerufen wird.
* reset() : Löscht das Gezeichnete. Setzt die Einstellungen auf den Standard zurück.
* clear() : Löscht das Gezeichnete. Der Turtle bleibt unverändert.
* write() : Schreibt einen Text.

Weitere Funktionen unter: https://docs.python.org/3/library/turtle.html


#### Beispiel-Code
Im Folgenden ist ein Beispiel angeführt.

In diesem Beispiel wird eine Instanz der Klasse Turtle erstellt. Es kann pro Zeichenfläche mehrere Turtles geben. Alle Operationen werden auf der Turtle-Instanz ausgeführt. Es gibt auch den Modus, ohne Instanzen zu arbeiten, dann wird mit einem *Default*-Turtle gearbeitet. Um den Umgang mit Objekten zu üben, verwenden Sie bitte die Instanz eines Turtles.

```python
#Kopieren Sie diesen Beispiel-Code in eine *.py-Datei.
#Das Beispiel stammt von der Python-Dokumentation zur Turtle-Graphic.

from turtle import *

t = Turtle()
t.color('red', 'yellow')
t.speed(0)
t.begin_fill()
while True:
    t.forward(200)
    t.left(170)
    if abs(t.pos()) < 1: break
t.end_fill()
done()
```

### Aufgabe 1: Zweidimensionale L-Systeme

Bei L-Systeme (benannt nach dem Biologen Aristid Lindenmayer) handelt es sich um einen mathematischen Formalismus. L-Systeme wurden als Grundlage einer axiomatischen Theorie biologischer Entwicklung vorgeschlagen und werden auch verwendet, um Fraktale zu generieren.

Die zentrale Idee der L-Systeme besteht darin, Ersetzungsregeln im Sinne von Grammatiken rekursiv anzuwenden. Dadurch entstehen aus einfachen Symbolketten (genannt Axiome) beliebig lange und komplizierte Symbolketten.

Wenn man sich bei den Symbolen auf F, B, + und - beschränkt und diese im Sinne einer *Turtle*-Graphics zum Bewegen eines virtuellen Zeichenstifts wie folgt interpretiert:
* *F* bewegt den Zeichenstift um *x*-Einheiten nach vorwärts (engl. forward),
* *B* bewegt den Zeichenstift um *x*-Einheiten nach rückwärts (engl. backward),
* *+* ändert die Richtung des Zeichenstifts um *t* Grad nach links und
* *-* ändert die Richtung des Zeichenstifts um *t* Grad nach rechts,
dann lassen sich aus den resultierenden Symbolketten Grafiken erzeugen.

Beispiel: Mit der Ersetzungsregel *F = F + F - - F + F*, dem Axiom F, einer Schrittweite s >= 1 und einem Winkel t = 60 Grad, ergibt sich durch sukzessive Anwendung:

* Schritt 0, Axiom: *F*
* Nach Schritt 1: *F + F - - F + F*
* Nach Schritt 2: *F + F - - F + F + F + F - - F + F - - F + F - - F + F + F + F - - F + F*
